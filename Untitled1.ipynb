```python
# ============================================================
# 🏠 Nepal Housing Price Analysis
# WorldQuant University — Applied Data Science Lab
# Project 1: Housing in Mexico (adapted to Nepal context)
# Lesson: Tabular and Tidy Data
# By: Priyotam
# ============================================================

# ------------------------------------------------------------
# SECTION 1: Working with Lists
# The simplest way to store a single house's data
# Problem: we can't tell what each number means just by looking
# ------------------------------------------------------------

# A single house: [price in NPR, area in sqm, number of rooms]
house_0 = [8500000, 120, 3]

# Task 1.1.1 — Calculate price per square meter
# Dividing price by area lets us compare houses of different sizes fairly
price_per_sqm = house_0[0] / house_0[1]
print("Price per sq meter: NPR", price_per_sqm)

# Task 1.1.2 — Add price per sqm back into the house data
house_0.append(price_per_sqm)
print("Updated house data:", house_0)

# ------------------------------------------------------------
# SECTION 2: Working with Nested Lists
# Multiple houses stored as a list of lists
# ------------------------------------------------------------

nepal_houses = [
    [8500000, 120, 3],   # Kathmandu
    [4200000, 85, 2],    # Pokhara
    [2100000, 60, 2],    # Butwal
    [3800000, 95, 3],    # Chitwan
    [6700000, 110, 4],   # Lalitpur
]

# Task 1.1.3 — Use a loop to calculate price per sqm for all houses
# Much better than doing it manually for each house!
for house in nepal_houses:
    price_per_sqm = house[0] / house[1]
    house.append(price_per_sqm)

print("Houses with price per sqm added:")
nepal_houses

# ------------------------------------------------------------
# SECTION 3: Working with Dictionaries
# Better than lists — each value now has a label (key)
# Much easier to understand what each value represents
# ------------------------------------------------------------

# Single house as a dictionary
house_0_dict = {
    "price_npr": 8500000,
    "area_sqm": 120,
    "rooms": 3,
}

# Task 1.1.4 — Add price per sqm to the dictionary
house_0_dict["price_per_sqm"] = house_0_dict["price_npr"] / house_0_dict["area_sqm"]
print("House with price per sqm:", house_0_dict)

# ------------------------------------------------------------
# SECTION 4: List of Dictionaries (Row-wise)
# Each dictionary = one house (one row)
# Easy for row-wise operations, harder for column-wise
# ------------------------------------------------------------

nepal_houses_rowwise = [
    {"price_npr": 8500000, "area_sqm": 120, "rooms": 3},
    {"price_npr": 4200000, "area_sqm": 85,  "rooms": 2},
    {"price_npr": 2100000, "area_sqm": 60,  "rooms": 2},
    {"price_npr": 3800000, "area_sqm": 95,  "rooms": 3},
    {"price_npr": 6700000, "area_sqm": 110, "rooms": 4},
]

# Task 1.1.5 — Add price per sqm to every house using a loop
for house in nepal_houses_rowwise:
    house["price_per_sqm"] = house["price_npr"] / house["area_sqm"]

nepal_houses_rowwise

# Task 1.1.6 — Calculate mean house price (row-wise is harder for this)
house_prices = []
for house in nepal_houses_rowwise:
    house_prices.append(house["price_npr"])

mean_price = sum(house_prices) / len(house_prices)
print("Mean house price: NPR", mean_price)

# ------------------------------------------------------------
# SECTION 5: Dictionary of Lists (Column-wise)
# Each key = one feature, value = list of all house values
# Easier for column-wise operations like mean, sum etc.
# ------------------------------------------------------------

nepal_houses_columnwise = {
    "price_npr":  [8500000, 4200000, 2100000, 3800000, 6700000],
    "area_sqm":   [120, 85, 60, 95, 110],
    "rooms":      [3, 2, 2, 3, 4],
}

# Task 1.1.7 — Calculate mean house price (much easier column-wise!)
mean_price = sum(nepal_houses_columnwise["price_npr"]) / len(nepal_houses_columnwise["price_npr"])
print("Mean house price: NPR", mean_price)

# Task 1.1.8 — Add price per sqm column using zip
# zip() lets us loop through two lists at the same time
price_per_sqm_list = []
for p, a in zip(nepal_houses_columnwise["price_npr"], nepal_houses_columnwise["area_sqm"]):
    price_per_sqm_list.append(p / a)

nepal_houses_columnwise["price_per_sqm"] = price_per_sqm_list
nepal_houses_columnwise

# ------------------------------------------------------------
# SECTION 6: Pandas DataFrame
# The best way to work with tabular data in Python
# Combines the best of both row-wise and column-wise operations
# ------------------------------------------------------------

import pandas as pd

df_houses = pd.DataFrame(nepal_houses_columnwise)

print("Shape:", df_houses.shape)
df_houses
```
